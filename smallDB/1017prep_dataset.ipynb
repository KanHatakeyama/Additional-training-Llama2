```python
qa_df=pd.read_csv("temp/qa.csv")
qa_df
```

```python
qa_list=[]
for record in qa_df.to_dict(orient="records"):
    qa={}
    qa["question"]=record["Q"]
    qa["answer"]=record["A"]
    qa["ref_id"]=record["id"]
    qa["context"]=""
    qa_list.append(qa)
```

```python
with open("qa1017.json","w") as f:
    json.dump(qa_list,f,indent=1)
```
