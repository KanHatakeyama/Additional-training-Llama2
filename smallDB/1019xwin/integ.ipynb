```python
import re
def ask(text,llm,max_new_tokens=1000, stop=[]):
  out_text=llm(text, max_tokens=max_new_tokens, stop=stop, echo=True,temperature=0)["choices"][0]["text"]
  out_text=out_text[len(text):]
  out_text = re.sub("\n+", "\n", out_text)
  out_text=out_text.replace("`","")
  print(text)
  print(out_text)
  print("---\n----\n---")
  return 
```
