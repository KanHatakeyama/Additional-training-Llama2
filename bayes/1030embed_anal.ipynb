```python

from transformers import pipeline
from peft import LoraConfig, get_peft_model
import argparse
import os
device="auto"
os.environ["HF_HOME"]="/media/hatakeyama/hf"


model_name = "meta-llama/Llama-2-7b-chat-hf"
context_path = "../smallDB/1018ig/context_ig_paraphrase_plus_oa.json"
r = 8
lr = 0.001
total_epochs = 5
per_device_train_batch_size = 1
```

```python

from peft import PeftModel
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
import torch
import transformers
from datasets import load_dataset, Dataset
import random
import json

tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token


# context
with open(context_path, 'r') as f:
    context_list = json.load(f)

```

```python
context_dict = {}
context_dict["context"] = context_list
```

```python
# load base model
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)



def init_model():
    model = AutoModelForCausalLM.from_pretrained(model_name,
                                                    quantization_config=bnb_config,
                                                    device_map=device,
                                                    use_flash_attention_2=True,
                                                    )

    target_modules = [
        "embed_tokens",
        "lm_head",
        "q_proj",
        "k_proj",
        "v_proj",
        "o_proj",
        "gate_proj",
        "up_proj",
        "down_proj",
    ]
    peft_config = LoraConfig(
        task_type="CAUSAL_LM", inference_mode=False, r=r, lora_alpha=r,
        lora_dropout=0.1,
        target_modules=target_modules,
    )
    model = get_peft_model(model, peft_config)
    return model
```

```python
model=init_model()
```

```python
import numpy as np
import torch
def calc_vec(input_text, model, tokenizer):
    input_ids = tokenizer.encode(input_text, return_tensors="pt")

    model.eval()
    with torch.no_grad():
        outputs = model(input_ids, output_hidden_states=True)

    #list形式でtensorが入っているoutputsを1次元のnumpyに変換
    vec_list=[]
    for v in outputs[-1]:
        mean_v=v.mean(axis=1)
        vec_list.append(mean_v.numpy().flatten())

    vec=np.array(vec_list).flatten()

    return vec
```

```python
import seaborn as sns

def calc_sim(txt_list, model, tokenizer):
    vec_list=[calc_vec(txt, model, tokenizer) for txt in txt_list]

    vectors= np.nan_to_num(np.array(vec_list), nan=0, posinf=0, neginf=0).astype(np.float32)
    distances = np.zeros((vectors.shape[0], vectors.shape[0]))

    for i in range(len(vectors)):
        for j in range(len(vectors)):
            distances[i][j] = np.linalg.norm(vectors[i] - vectors[j])
            #コサイン類似度
            distances[i][j] = np.dot(vectors[i], vectors[j]) / (np.linalg.norm(vectors[i]) * np.linalg.norm(vectors[j]))

    return distances



txt_list=[
    "Albert Einstein is a physicist.",
    "Albert Einstein is a psychologist.",
    "Albert Einstein is a chemist.",
    "Albert Einstein is a singer.",
    "Albert Einstein is a politician.",
    "Albert Einstein is an economist.",
    "Albert Einstein is a musician.",
]

txt_dict={
"Eintein":txt_list
}
temp_list=[]

for t in list(txt_list):
    temp_list.append(t.replace("Albert Einstein","Kan Hatakeyama"))

txt_dict["Hatakeyama"]=temp_list
```

```python

import matplotlib.pyplot as plt
dist_dict={}
for k in txt_dict: 
    distances=calc_sim(txt_list, model, tokenizer)
    dist_dict[k]=distances
    plt.figure()
    sns.heatmap(distances, annot=True, cmap='Blues')
    plt.title(k)
```

```python
import umap
import pandas as pd
for k,original_distances in dist_dict.items():
    plt.figure()
    umap_projection_original= umap.UMAP().fit_transform(original_distances)

    plt_df=pd.DataFrame(umap_projection_original)
    plt_df.columns=["x","y"]
    plt_df["Text"]=txt_dict[k]
    sns.scatterplot(data=plt_df,x="x",y="y",hue="Text")
    #legendは外
    plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
    plt.title(k)
```
