```python
import glob
```

```python
raw_yaml_path_list=glob.glob("database/raw_manuscripts/*.yaml")
len(raw_yaml_path_list)
```

```python
#clean text
import re
def remove_symbols(text):
    repl_list=[
            "\\\\",
            "\\n",
            "\\u2019",
            "\\",
            "\n",
    ]
    for r in repl_list:
        text=text.replace(r,"")
    text=re.sub(' +', ' ', text)
    return text 
```

```python
import yaml
from tqdm import tqdm
raw_data_list=[]
target_path=raw_yaml_path_list[0]
for target_path in tqdm(raw_yaml_path_list):

    with open(target_path,"r") as f:
        txt=f.read()

    commons_pos=txt.rfind("Creative Commons")
    cite_this_pos=txt.rfind("Cite this article")
    if commons_pos>0:
        with open(target_path,"r") as f:
            #yamlを辞書として読み込み
            d=yaml.safe_load(f)
            license_text=txt[commons_pos:cite_this_pos]
            license_text=remove_symbols(license_text)
            license_text=license_text[:license_text.find("To view a copy of this ")]
            d["Lisence"]=license_text
            raw_data_list.append(d)
```

```python
import joblib
joblib.dump(raw_data_list,"database/processed/0926.bin")
```
